```python
import os
from os.path import isdir
import cv2
import numpy as np

foldername1 = 'train'
foldername2 = 'train'

path = os.path.join(os.getcwd(), foldername1, foldername2)

memes = []

for folder in os.listdir(path):
  if isdir(os.path.join(path, folder)): 
    for file in os.listdir(os.path.join(path, folder)):                              # get every file in the folder
      img = cv2.imread(os.path.join(path, folder, file))                  # load the image
      gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)             # make it RGB (cv2 uses BGR)
      res = cv2.resize(gray, dsize=(100, 100))                   #resize it so all images are one size
      memes.append(res)


foldername3 = 'valid'
foldername4 = 'valid'
path = os.path.join(os.getcwd(), foldername3, foldername4)

for folder in os.listdir(path):
  if isdir(os.path.join(path, folder)): 
    for file in os.listdir(os.path.join(path, folder)):                              # get every file in the folder
      img = cv2.imread(os.path.join(path, folder, file))                  # load the image
      gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)             # make it RGB (cv2 uses BGR)
      res = cv2.resize(gray, dsize=(100, 100))                   #resize it so all images are one size
      memes.append(res)

memes = np.array(memes)
memes.shape

```

```python
X = []

foldername = 'InstaNY100K/img_resized/newyork'

path = os.path.join(os.getcwd(), foldername)                    #define foldername

i = 0

for file in os.listdir(path):                                   # get every file in the folder
    i += 1
    img = cv2.imread(os.path.join(path, file))                   # load the image
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)             # make it RGB (cv2 uses BGR)
    res = cv2.resize(gray, dsize=(100, 100))                   #resize it so all images are one size
    X.append(res) 
    if i >= 5000:                                               # number of images to load
      break

not_memes = np.array(X)
not_memes.shape
```

```python
memes = memes[:5000]
not_memes = not_memes[:5000]

memes.shape = not_memes.shape
```

```python
y1 = np.array([[0] for x in range(5000)])

y2 = np.array([[1] for x in range(5000)])
```

```python
y_train = np.concatenate((y1[:4000], y2[:4000]))

y_test = np.concatenate((y1[4000:], y2[4000:]))

y_train.shape, y_test.shape
```

```python
X_train = np.concatenate((not_memes[:4000], memes[:4000]))

X_test = np.concatenate((not_memes[4000:], memes[4000:]))

X_train.shape, X_test.shape
```

```python
X_train_scaled = X_train / 255

X_test_scaled = X_test / 255
```

```python
X_train_scaled = np.expand_dims(X_train_scaled, axis = -1)
X_train_scaled.shape
```

```python
X_test_scaled = np.expand_dims(X_test_scaled, axis = -1)
X_test_scaled.shape
```

```python
new_random_train_indices = np.random.permutation(len(X_train_scaled))
X_train_reshuffled = X_train_scaled[new_random_train_indices]
y_train_reshuffled = y_train[new_random_train_indices]
```

```python
y1 = np.array([[0] for x in range(5000)])

y2 = np.array([[1] for x in range(5000)])


y_train = np.concatenate((y1[:4000], y2[:4000]))

y_test = np.concatenate((y1[4000:], y2[4000:]))

y_train.shape, y_test.shape

X_train = np.concatenate((not_memes[:4000], memes[:4000]))

X_test = np.concatenate((not_memes[4000:], memes[4000:]))

X_train.shape, X_test.shape
```

```python
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential

def initialize_model():
    model = Sequential()
    # layer 1
    model.add(layers.Conv2D(16, (4,4), input_shape=(100,100,1), 
                            padding='same', activation='relu'))
    model.add(layers.MaxPool2D(pool_size=(2,2)))
    # layer 2
    model.add(layers.Conv2D(32, (3,3), padding='same', activation='relu'))
    model.add(layers.MaxPool2D(pool_size=(2,2)))
    # layer 3
    model.add(layers.Conv2D(32, (3,3), padding='same', activation='relu'))
    model.add(layers.MaxPool2D(pool_size=(2,2)))
    # layer 4
    model.add(layers.Conv2D(32, (2,2), padding='same', activation='relu'))
    model.add(layers.MaxPool2D(pool_size=(2,2)))
    # flatten layer
    model.add(layers.Flatten())
    # Dense layer
    model.add(layers.Dense(10, activation='relu'))
    # last layer
    model.add(layers.Dense(1, activation='sigmoid'))
    # compile
    model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy']
                 )
    return model   
```

```python
model = initialize_model()

model.summary()
```

```python
# YOUR CODE HERE
from tensorflow.keras.callbacks import EarlyStopping

es = EarlyStopping(patience = 5, restore_best_weights= True)

# To complete
history = model.fit(X_train_reshuffled, y_train_reshuffled, validation_split = 0.2,
          epochs=200,  # Use early stop in practice
          batch_size=16, 
          verbose=1,
         callbacks = [es])

```

```python
model.evaluate(X_test_scaled, y_test)
```
